#Q3.

Mutable Objects are objects whose contents can be changed after they are created without changing their identity (their "address" in the computer's memory).

example : Lists, Dictionaries, Sets

Immutable Objects:
These are objects whose contents cannot be changed once they are created. If you try to modify an immutable object, Python actually creates a brand-new object in memory with the new value.

example: Integers, Floats, Strings, Tuples

#Q4.

Operators are special symbols in Python that carry out arithmetic or logical computations. The value that the operator operates on is called the operand.

In Python, operators are categorized into several groups based on the type of operation they perform.

###Arithmetic Operators
These are used to perform common mathematical operations like Addition, Subtraction, Multiplication, Division, etc.

###Comparison (Relational) Operators
These operators compare two values and return a Boolean value (True or False), like
Equal to (==): 5 == 5 → True

Not equal to (!=): 5 != 3 → True

Greater than (>): 5 > 10 → False

Less than or equal to (<=): 3 <= 3 → True

###Logical Operators
Logical operators are used to combine conditional statements. They are the "decision-makers" in your code

###Assignment Operators
These are used to assign values to variables. The most common is =, but Python offers "shortcut" operators that perform an operation and assignment at the same time.

###Membership Operators
These are unique to Python and are very readable. They test if a sequence (like a string, list, or tuple) is present in an object.

###Identity Operators
These compare the objects, not just if they are equal, but if they are actually the same object with the same memory location.

#Q5.

Type Casting is the process of converting a variable from one data type to another, There are two main types of type casting: Implicit and Explicit.

###Implicit Type Casting (Automatic)
Python automatically converts one data type to another without any user involvement. This usually happens to avoid data loss. For example, if you add an integer to a float, Python converts the integer to a float first.

###Explicit Type Casting (Manual)
This is when the programmer manually converts the data type using built-in functions. This is the most common form of type casting you will use.

Conditional statements are the decision makers of a program. They allow your code to execute specific blocks of instructions only when certain conditions are met. if a condition is True, the program goes one way; if it is False, it goes another.

In Python, we use three main keywords to handle these decisions: if, elif, and else.

In Python, loops are used to repeat a block of code multiple times. This is essential for tasks like processing items in a list, or performing calculations until a specific result is reached.

Python primarily uses two types of loops: for loops and while loops.

### for loop
The for loop is used for iterating over a sequence (like a list, a tuple, a dictionary, a set, or a string)

###while loop
The while loop repeats a block of code as long as a certain condition is True, generally we sue it When we don't know exactly how many times you need to repeat the task, and you are waiting for a specific state or user input.

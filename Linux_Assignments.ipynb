2. What is the difference between Hard Link & Soft Link?
    - A Hard Link is an additional name for an existing file. When you create a hard link, you are creating a second directory entry that points directly to the file's data on the hard drive.

      If you delete the original file, the data remains accessible through the hard link. The file is only deleted when all hard links pointing to its inode are removed.

    - A Soft Link is like a shortcut or an alias. It is a small file that simply stores the path (the name and location) of the original file.

      If the original file is renamed, moved, or deleted, the soft link will break because the path it's pointing to no longer exists.

3. What is a Kernel in Linux?
    - The Kernel is the most important piece of software in the Linux operating system. It's the core program that manages everything on your computer.

      It acts as a translator between all our applications (like web browser) and physical components (like CPU, RAM, disk). Browser doesn't know how to talk to a hard drive directly, so it asks the Kernel to do the work.
